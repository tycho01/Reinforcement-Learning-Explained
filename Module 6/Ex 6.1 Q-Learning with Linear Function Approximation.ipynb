```python
class Agent(object):  
        
    def __init__(self, actions):
        self.actions = actions
        self.num_actions = len(actions)

    def act(self, state):
        raise NotImplementedError
```

```python
class QLearningFAAgent(Agent):
    def __init__(self, actions, obs_size, epsilon=0.01, alpha=0.5, gamma=1):
        super(QLearningFAAgent, self).__init__(actions)
        
        ## TODO 1
        ## Initialize thetas here
        ## In addition, initialize the value of epsilon, alpha and gamma
        self.theta = np.zeros(obs_size * len(actions)) # , dtype=np.float64
        self.epsilon = epsilon
        self.alpha = alpha
        self.gamma = gamma
 
    def featureExtractor(self, state, action):
        feature = None
        
        actionindex = np.zeros(self.num_actions, dtype=np.int)
        actionindex[action] = 1
        feature = np.concatenate([actionindex[i] * state for i in self.actions])
        return feature    
            
    def act(self, state):
        ## epsilon greedy policy 
        if np.random.random() < self.epsilon:
            i = np.random.randint(0,len(self.actions))
        else:
            # q = [np.sum(self.theta.transpose() * self.featureExtractor(state, a)) for a in self.actions]

            ## TODO 2
            # Q(s, a) for a in all actions
            q = [np.sum(self.theta.transpose() * self.featureExtractor(state, a)) for a in self.actions]
            
            if q.count(max(q)) > 1:
                best = [i for i in range(len(self.actions)) if q[i] == max(q)]
                i = np.random.choice(best)
            else:
                i = q.index(max(q))
            
        action = self.actions[i]
        return action           
            
    def learn(self, state1, action1, reward, state2, done):
        
        """
        Q-learning with FA 
        theta <- theta + alpha * td_delta * f(s,a)
        where 
        td_delta = reward + gamma * max(Q(s') - Q(s,a))
        Q(s,a) = thetas * f(s,a)
        max(Q(s')) = max( [ thetas * f(s'a) for a in all actions] )
        
        """
        ## TODO 3
        ## Implement the q-learning update here
        max_idx = 0
        q_sum = [np.sum(self.theta.transpose() * self.featureExtractor(state2, a)) for a in self.actions]
        
        if q_sum.count(max(q_sum)) > 1:
            best = [i for i in range(len(self.actions)) if q_sum[i] == max(q_sum)]
            max_idx = np.random.choice(best)
        else:
            max_idx = q_sum.index(max(q_sum))
        # replace 0 with the correct calculation: max Q(s',a) for a in all actions
        maxqnew = np.sum(self.theta.transpose() * self.featureExtractor(state2, self.actions[max_idx]))
        # replace 0 with the correct calculation: Q(s,a) where a is the action taken
        oldv = np.sum(self.theta.transpose() * self.featureExtractor(state1, action1))

        td_target = reward + self.gamma * maxqnew
        td_delta = td_target - oldv
        self.theta += (self.alpha * td_delta * self.featureExtractor(state1, action1)) # replace 0 with the correct calculation -- thetas update equation
```
